```python
'''
Steve Winnall 10 Aug 2019
This code is work toward the Udacity Machine Learning NanoDegree Capstone Project
"The Art Thief - Is this a Whiteley?"
Using machine learning to identify a "fake" Brett Whiteley paiting.
This code is to graph the Receiver Operating Characteristic (ROC) and Area Under the Curve (AUC)
Useful for determining the True Positive Rate (TPR) against False Positive Rate (FPR) to determine discrimination ability of a model

'''

#imports
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#data 

''' #color
TPR= [0.94, 0.83, 0.79, 0.76, 0.55, 0.1]
FPR = [0.97, 0.75, 0.54, 0.45, 0.19, 0.09]
TPR = TPR[::-1] #reverse for integration
FPR = FPR[::-1] # reverse for integration''' 


#intensity
TPR = [ 0.98, 0.92, 0.85,  0.79, 0.76, 0.7, 0.55, 0.37, 0.05, 0.05 ]
FPR = [ 0.97, 0.79,  0.63, 0.56, 0.49, 0.4, 0.28, 0.14, 0.07, 0.05] 
TPR = TPR[::-1] #reverse for integration
FPR = FPR[::-1] # reverse for integration

chancex = np.linspace(0,1,10) #probability of chance 
chancey = chancex

# This is the ROC curve

plt.title('Receiver Operating Characteristic (ROC)')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.plot(chancex,chancey,'b--')
plt.plot(FPR,TPR,'ro')
plt.plot(FPR,TPR,'r')
plt.show()

# This is the AUC

auc = np.trapz(TPR,FPR)

print("----")
print("Area Under the ROC Curve (AUC): {:.3f}".format(auc))
print("----")
```
